```python
matrices = [generate_S("random", 5) for _ in range(100000)]
conds = [np.linalg.cond(S) for S in matrices]
```

```python
bins = np.logspace(np.log10(min(conds)), np.log10(max(conds)), 50)

plt.hist(conds, bins=bins, edgecolor='black')
plt.xscale('log')
plt.title('Distribution of Condition Numbers for Random Matrices')
```

```python
x = np.pow(10, np.linspace(np.log10(min(conds)/5), np.log10(max(conds)/5), 100))
fig, ax1 = plt.subplots()

# First plot — histogram
ax1.hist(conds, edgecolor='black', bins=bins, cumulative=False)
ax1.set_xscale('log')
ax1.set_ylabel('Cumulative count')

# Create second axis
ax2 = ax1.twinx()

# Second plot — function on the same x axis
y = (2 * x + 4) / np.pow(x, 3) * np.exp(-2/x - 2/x**2)
ax2.plot(x * 5, y, color='red', label='Edelman', marker='x')
# Optional: legend
fig.legend(loc='upper right')

plt.show()
```
