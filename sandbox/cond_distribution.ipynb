```python
from sandbox.plt_dict import matplotlib_update
import numpy as np
import matplotlib.pyplot as plt

matplotlib_update()

np.random.seed(213)
```

```python
def f(x):
    return (2*x+4)/(x**3) * np.exp(-2/x - 2/(x**2))
```

```python
n_matrices = 10_000
d = 30 
xmax = 25

kappas = []

for _ in range(n_matrices):
    A = np.random.randn(d,d)
    kappa_a = np.linalg.cond(A)
    kappas.append(kappa_a)

kappas = np.array(kappas)/d
plt.hist(kappas, density=True, alpha=0.3, bins=np.linspace(0, xmax, 50))
plt.plot(x := np.linspace(0.001, xmax, num=1000), f(x), color="red")
plt.ylabel("Density")
plt.xlabel("$\\kappa(A)/d$")
plt.xlim(0, xmax)
plt.grid(alpha=0.7)
plt.savefig("latex/images/edelman.pdf")
plt.show()
```
