```python
import numpy as np
import matplotlib.pyplot as plt
from sandbox.jordan8 import check_doubling_consistency, get_superdiagonal, per_power_features
```

```python
def test():
    J = np.diag(get_superdiagonal(10, 35), k=1)
    S = np.random.randn(35, 35)
    E = np.random.randn(35, 35)/np.sqrt(35)

    A = np.linalg.inv(S) @ (J + 1e-5 * E) @ S

    A /= np.max(np.abs(np.linalg.eigvals(A)))

    P = per_power_features(A)

    return np.max(check_doubling_consistency(P, verbose=True)[0])
```

```python
results = [test() for _ in range(1000)]
```

```python
plt.hist(np.log(results)/np.log(10))
plt.plot()
```
