```python
# Regex to extract values
pattern = re.compile(
    r"Epoch \[(\d+)/\d+\] \| Train Loss: ([\d.]+) \| Val Loss: ([\d.]+) \| LR: ([\deE\-.]+)"
)

output = StringIO()
writer = csv.writer(output)

# Write header
writer.writerow(["epoch", "train_loss", "val_loss", "lr"])

# Parse lines
for line in log_text.strip().splitlines():
    match = pattern.search(line)
    if match:
        epoch, train_loss, val_loss, lr = match.groups()
        writer.writerow([int(epoch)-1, train_loss, val_loss, lr])

# Get CSV string
csv_string = output.getvalue()
print(csv_string)
```
