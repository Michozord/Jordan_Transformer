```python
import numpy as np
d = 5
```

```python
J = 1 * np.eye(d) + np.diag([1] * 4 + [0] * (d - 4 - 1), k=1)
```

```python
J
```

```python
S = np.random.rand(d,d)
```

```python
A = np.linalg.inv(S) @ J @ S
```

```python
A 
```

```python
import sympy as sp
```

```python
A_sp = sp.Matrix(A)
```

```python
A_sp
```

```python
J, P = A_sp.jordan_form()
```
