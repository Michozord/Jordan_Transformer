```python
import torch
import torch.nn as nn
from jordan_transformer.jordan import MatrixEncoder, JordanClassifier, JordanTransformer

def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# Testing with d=5, m=32
d, m = 5, 32
encoder = MatrixEncoder(d, out_dim=m)
classifier = JordanClassifier(num_classes=d, in_dim=m)
transformer = JordanTransformer(dim=m)

print(f"--- Verification for d={d}, m={m} ---")
print(f"MatrixEncoder: {count_parameters(encoder)}")
print(f"JordanClassifier: {count_parameters(classifier)}")
print(f"JordanTransformer: {count_parameters(transformer)}")
```
